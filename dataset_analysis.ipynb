```python
retail.drop_duplicates(inplace=True)
```

```python
# we have 4382 missing description in rows. As we are dealing with around 10L rows over two years it is safe to drop them.
retail['description'][retail['description']==None]
```

```python
retail.dropna(subset=['description'],inplace=True)
```

```python
# now we have around 243k rows with missing customer_id ,let's assume these were purchases without customer account . Lets call them Guest and fill these rows with customer_id=0
retail['customer_id']=retail['customer_id'].fillna(0)
```

```python
retail['invoicedate'] = pd.to_datetime(retail['invoicedate'])
retail['year']=retail['invoicedate'].dt.year
retail['month']=retail['invoicedate'].dt.month
```

```python
retail['total_revenue']=retail['quantity'] *  retail['price']
```

```python
#total revenue also contains when the quantity is negative 
#only positive
gross_revenue=retail[retail['quantity']>0]['total_revenue'].sum()
print(gross_revenue)
#negative also included
net_revenue=retail['total_revenue'].sum()
print(net_revenue)

gap=gross_revenue-net_revenue
print(gap)
```
